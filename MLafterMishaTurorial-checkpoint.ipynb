```python
from rdkit.Chem import AllChem, DataStructs
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
```

```python
model = load_model(r"C:\Users\elseh\Downloads\ic50_func.h5")
```

```python
bits = 1024
smiles = "CCOc1cc(NC)ccc1C4CC(=O)Nc3cc2OCOc2cc34"
m1 = AllChem.MolFromSmiles(smiles)
ecfp = AllChem.GetHashedMorganFingerprint(m1, 3, nBits=bits)
fp = np.zeros((1024, 1024), dtype=np.int8)
# fp = np.zeros((0,), dtype=np.int8)
DataStructs.ConvertToNumpyArray(ecfp, fp)
g = pd.read_csv(r'C:\Users\elseh\Downloads\genomic.csv')
```

```python
fp = np.asarray(fp).astype(np.float32)
```

```python
bitstring="".join(fp.astype(str))
fp2 = DataStructs.cDataStructs.CreateFromBitString(bitstring)
# print(*fp2, sep='\n')
# print(*fp, sep='\n')
```

```python
input_1 = np.zeros((1024,1024))
input_1[0] = np.array(fp)
input_2 = g.iloc[0:1024, 1:-2]
```

```python
# ||||||THIS WORKS||||||
prediction = model.predict([input_1, input_2], batch_size=None, verbose=1, steps=None) 
# prediction = model.predict([fp, input_2], batch_size=None, verbose=1, steps=None)
```
